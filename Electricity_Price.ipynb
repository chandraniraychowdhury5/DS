```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

```

```python
df = pd.read_csv('/content/drive/MyDrive/datasetAndProgram/renewable electricity by country - Electricity Prices Worldwide by country 2022.csv')
```

```python
print(df.shape)
print(df.head())
print(df.info())
print(df.columns)
df.rename(columns={'household price $/kWh': 'householdprice', 'business price $/kWh': 'businessprice', 'average price $/kWh': 'averageprice'}, inplace = True)
print(df.columns)
```

```python
print(df.describe())
print(df.columns)
```

```python
print(df.nunique())
```

```python
print(df.sample(10))
```

```python
print(df.isna().sum())
```

```python
plt.figure(figsize=(10,5))
plt.hist(df['averageprice'], bins = 20, color = "red", edgecolor = "green")
plt.xlabel("Average Price")
plt.ylabel("Frequency")
plt.title("Histogram of Average Price")
plt.plot()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(df['householdprice'], bins = 20, color = "green", edgecolor = "black")
plt.xlabel("Household Price")
plt.ylabel("Frequency")
plt.title("Histogram of Household Price")
plt.plot()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(df['businessprice'], bins = 20, color = "blue", edgecolor = "green")
plt.xlabel("Business Price")
plt.ylabel("Frequency")
plt.title("Histogram of Business Price")
plt.plot()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(df["Country"], df["householdprice"], color = "green", marker = "*")
plt.xlabel("Country")
plt.ylabel("Household Price")
plt.title("Scatter plot for Household Price")
plt.plot()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(df["Country"], df["averageprice"], color = "red", marker = "o")
plt.xlabel("Country")
plt.ylabel("Average Price")
plt.title("Scatter plot for Average Price")
plt.plot()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(df["Country"], df["businessprice"], color = "blue", marker = "*")
plt.xlabel("Country")
plt.ylabel("Business Price")
plt.title("Scatter plot for Business Price")
plt.plot()
plt.show()
```

```python
plt.figure(figsize=(10,5))
x = pd.Categorical(df["Country"]).codes
sns.regplot(data = df, x = x, y = df["averageprice"], color = "red")
plt.xlabel("Country")
plt.ylabel("Average Price")
plt.title("Regression plot for Average Price")
plt.plot()
plt.show()
```

```python
alldata = [df["averageprice"], df["householdprice"]]
labels = ["averageprice", "householdprice"]
fig, ax = plt.subplots(figsize = (8, 8))
ax.boxplot(alldata, labels)
ax.set_xlabel("Categories")
ax.set_ylabel("Value")
ax.set_title("Box plots")
plt.plot()
plt.show()
```
